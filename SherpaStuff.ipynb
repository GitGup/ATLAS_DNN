```python
dfs = []
count = 0
for index, row in Sherpa_df.iterrows():
    
    if (count == 100000):
        break
    pts = row['DMesons_pt']
    etas = row['DMesons_eta']
    ms = row['DMesons_m']
    lxys = row['DMesons_fitOutput__Lxy']
    costhetastars = row['DMesons_costhetastar']
    pdgids = row["DMesons_pdgId"]
        
    d = {Sherpa_df.columns[0]: pts, Sherpa_df.columns[1]: etas, Sherpa_df.columns[2]: ms, Sherpa_df.columns[3]: lxys, 
         Sherpa_df.columns[4]: costhetastars, Sherpa_df.columns[-1]: pdgids}
    df = pd.DataFrame(data=d)
    
    dfs.append(df)
    count+=1
```

```python
# df = dfs[0].append(dfs[1])

df = pd.concat(dfs)
df = df.reset_index(drop = True)
df
```

```python
#func = lambda x: 'only_411' if x == [411] else ('only_4122' if x == [4122] else 'both')

def func(x):
    if x == 411:
        return 'only_411'
    elif x == 4122:
        return 'only_4122'
    elif x == -4122:
        return 'only_-4122'
    elif x == -411:
        return 'only_-411'
    else:
        return 'both'

df['particle_type'] = df['DMesons_pdgId'].apply(func)

# separate the rows based on the new column
only_411 = df[df['particle_type'] == 'only_411']
only_4122 = df[df['particle_type'] == 'only_4122']
only_411_negative = df[df['particle_type'] == 'only_-411']
only_4122_negative = df[df['particle_type'] == 'only_-4122']
#new_df = df[(df['particle_type'] != 'both') & (df['particle_type'] != 'only_-411')]

new_df = df[(df['particle_type'] != 'only_-411') & (df['particle_type'] != 'only_-4122')]
new_df = new_df.reset_index(drop=True)

new_df.loc[:, 'DMesons_pt'] = new_df['DMesons_pt'].apply(lambda x: float(x))
new_df.loc[:, 'DMesons_eta'] = new_df['DMesons_eta'].apply(lambda x: float(x))
new_df.loc[:, 'DMesons_m'] = new_df['DMesons_m'].apply(lambda x: float(x))
new_df.loc[:, 'DMesons_fitOutput__Lxy'] = new_df['DMesons_fitOutput__Lxy'].apply(lambda x: float(x))
new_df.loc[:, 'DMesons_costhetastar'] = new_df['DMesons_costhetastar'].apply(lambda x: float(x))
new_df.loc[:, 'DMesons_pdgId'] = new_df['DMesons_pdgId'].apply(lambda x: int(x))
new_df = new_df.drop('particle_type', axis=1)
new_df
```
